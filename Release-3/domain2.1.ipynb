```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/2023&2024domain.csv") # Replace with your actual file path
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
```

```python
df['Conversion Rate'] = df['Unique Clicks'] / df['Sent']
```

```python
conversion_summary = df.groupby(['Domain', 'Year'])['Conversion Rate'].mean().unstack()
```

```python
conversion_summary = df.groupby(['Domain', 'Year'])['Conversion Rate'].mean().unstack()
```

```python
conversion_summary.plot(kind='bar', figsize=(12, 6))
plt.title("Comparison of Conversion Rate by Domain (2023 vs 2024)")
plt.ylabel("Conversion Rate")
plt.xlabel("Email Domain")
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```
