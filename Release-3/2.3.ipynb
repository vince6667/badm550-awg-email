```python
import pandas as pd

# File path for the uploaded file
file_path = '2023&2024_newdomain_merged.xlsx'

# Load the Excel file
df = pd.read_excel(file_path)

# Show the first few rows of the DataFrame
print(df.head())
```

```python
# Display basic information about the DataFrame
print(df.info())

# Show summary statistics for numerical columns
print(df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the Conversion Rate for each domain
df['Conversion Rate'] = (df['Unique Clicks'] / df['Sent']) * 100

# Convert 'Date' to datetime format and extract the year
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year

# Filter for 2023 data
df_2023 = df[df['Year'] == 2023]

# Calculate the top 5 domains by conversion rate for 2023
df_2023_grouped = df_2023.groupby('Domain')['Conversion Rate'].mean()
top_5_2023 = df_2023_grouped.sort_values(ascending=False).head(5)

# Plot the graph for 2023
plt.figure(figsize=(10, 6))
top_5_2023.plot(kind='bar', color='skyblue')
plt.title('Top 5 Domains by Conversion Rate in 2023')
plt.xlabel('Domain')
plt.ylabel('Average Conversion Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Filter for 2024 data
df_2024 = df[df['Year'] == 2024]

# Calculate the top 5 domains by conversion rate for 2024
df_2024_grouped = df_2024.groupby('Domain')['Conversion Rate'].mean()
top_5_2024 = df_2024_grouped.sort_values(ascending=False).head(5)

# Plot the graph for 2024
plt.figure(figsize=(10, 6))
top_5_2024.plot(kind='bar', color='lightcoral')
plt.title('Top 5 Domains by Conversion Rate in 2024')
plt.xlabel('Domain')
plt.ylabel('Average Conversion Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
