2.7 Total Clicks over Total Views Comparison For Top Five Domains in 2023 and 2024




```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
data = pd.read_excel("Newsletters.xlsx")

# Extract the 'Year' from the 'Date' column
data['Year'] = pd.DatetimeIndex(data['Date']).year

# Filter the data for the years 2023 and 2024
data_filtered = data[data['Year'].isin([2023, 2024])]

# Find domains with data in both 2023 and 2024
domains_2023 = set(data_filtered[data_filtered['Year'] == 2023]['Domain'])
domains_2024 = set(data_filtered[data_filtered['Year'] == 2024]['Domain'])
common_domains = domains_2023.intersection(domains_2024)

# Filter data for only those domains present in both years
common_domain_data = data_filtered[data_filtered['Domain'].isin(common_domains)]

# Calculate Click Through Rate (CTR) by dividing Total Clicks by Total Views
common_domain_data['CTR'] = common_domain_data['Total Clicks'] / common_domain_data['Total Views']

# Group by 'Domain' and 'Year' to calculate total CTR for each domain in each year
domain_year_total_ctr = common_domain_data.groupby(['Domain', 'Year'])['CTR'].mean().reset_index()

# Find the top 5 domains based on the average CTR across both years
top_5_domains_avg_ctr = domain_year_total_ctr.groupby('Domain')['CTR'].mean().nlargest(5).index

# Filter data for only the top 5 domains
top_5_data_avg_ctr = domain_year_total_ctr[domain_year_total_ctr['Domain'].isin(top_5_domains_avg_ctr)]

# Pivot data for plotting, sorting in descending order of 2023's CTR
top_5_data_avg_ctr_pivot = top_5_data_avg_ctr.pivot(index='Domain', columns='Year', values='CTR')
top_5_data_avg_ctr_pivot = top_5_data_avg_ctr_pivot.sort_values(by=2023, ascending=False)  # Sort by 2023's values

# Plot the multi-bar chart for click-through rate (CTR)
fig, ax = plt.subplots(figsize=(10, 6))
top_5_data_avg_ctr_pivot.plot(kind='bar', ax=ax, color=['#1f77b4', '#ff7f0e'], width=0.8)
ax.set_title('Yearly Comparison of Total Clicks/Total Views for Top 5 Domains (2023 vs 2024)')
ax.set_xlabel('Domain')
ax.set_ylabel('Total Clicks/Total Views')  # Label updated for CTR
ax.set_xticklabels(top_5_data_avg_ctr_pivot.index, rotation=45)
ax.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```
