```python
import pandas as pd
# Load the dataset
file_path = "/content/2023&2024_newdomain_merged(Sheet1).csv"
df = pd.read_csv(file_path)
```

```python
print(df['Date'].dtype)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Year'] = df['Date'].dt.year
```

```python
print(df[['Date', 'Year']].head())
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df['Conversion Rate'] = (df['Unique Clicks'] / df['Sent']) * 100
```

```python
conversion_rate_comparison = df.groupby(['Domain', 'Year'])['Conversion Rate'].mean().unstack()
```

```python
# 6️⃣ 显示转换率对比结果
print(conversion_rate_comparison)
```

```python
import matplotlib.pyplot as plt

conversion_rate_comparison.plot(kind='bar', figsize=(12, 6))
plt.title("Comparison of Conversion Rate (2023 vs 2024) by Domain")
plt.ylabel("Conversion Rate (%)")
plt.xlabel("Domain")
plt.legend(title="Year")
plt.xticks(rotation=75)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
