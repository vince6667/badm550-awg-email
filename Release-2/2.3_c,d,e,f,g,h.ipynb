```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
file_path = 'Links.csv'  # Update with the actual file path
df = pd.read_csv(file_path)

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract year from the date
df['Year'] = df['Date'].dt.year
```

**2.3 c**

```python
# Find URLs that exist in both 2023 and 2024
df_2023 = df[df['Year'] == 2023]
df_2024 = df[df['Year'] == 2024]
common_urls = set(df_2023['URL']).intersection(set(df_2024['URL']))

# Filter dataset for common URLs
df_common = df[df['URL'].isin(common_urls)]

# Get top 5 common URLs based on total unique clicks over both years
top5_common = df_common.groupby('URL')['Unique Clicks'].sum().nlargest(5).reset_index()

# Plot for common URLs
plt.figure(figsize=(10, 5))
plt.bar(top5_common['URL'], top5_common['Unique Clicks'], color='green')
plt.xlabel('URL')
plt.ylabel('Total Unique Clicks (2023 & 2024)')
plt.title('Top 5 URLs Found in Both 2023 and 2024 with Highest Unique Clicks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**2.3 d**

```python

# Ensure top5_common is a proper DataFrame
top5_common = df_common.groupby('URL', as_index=False)['Unique Clicks'].sum().nlargest(5, 'Unique Clicks')

# Extract the list of top 5 URLs
top5_urls = top5_common['URL'].tolist()

# Filter dataset for only the top 5 common URLs
df_top5_common = df[df['URL'].isin(top5_urls)]

# Group by Year and URL to get unique clicks per year
df_plot = df_top5_common.groupby(['Year', 'URL'])['Unique Clicks'].sum().unstack()

# Plot multi-bar graph
bar_width = 0.4
x = np.arange(len(df_plot.columns))  # Positions for URLs

plt.figure(figsize=(10, 5))

# Plot bars for each year
for i, year in enumerate(df_plot.index):
    plt.bar(x + i * bar_width, df_plot.loc[year], width=bar_width, label=str(year))

plt.xlabel('URL')
plt.ylabel('Unique Clicks')
plt.title('Performance of Top 5 Common URLs in 2023 vs 2024')
plt.xticks(x + bar_width / 2, df_plot.columns, rotation=45, ha='right')
plt.legend(title='Year')
plt.show()
```
