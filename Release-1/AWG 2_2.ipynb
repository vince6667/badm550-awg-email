```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

file_path = "2023&2024_without_2023_nodate_files.csv"
df = pd.read_csv(file_path)

df['Date'] = pd.to_datetime(df['Date'])

df['% of Views'] = df['% of Views'].str.rstrip('%').astype(float)

df_2023 = df[df['Date'].dt.year == 2023]
df_2024 = df[df['Date'].dt.year == 2024]

df_2023_grouped = df_2023.groupby('Date').agg({
    'Unique Clicks': 'sum',
    '% of Views': 'mean'
}).reset_index()

df_2024_grouped = df_2024.groupby('Date').agg({
    'Unique Clicks': 'sum',
    '% of Views': 'mean'
}).reset_index()





plt.figure(figsize=(8, 6))
sns.kdeplot(x=df_2023_grouped['Unique Clicks'], y=df_2023_grouped['% of Views'], cmap="coolwarm", fill=True, alpha=0.7)
plt.scatter(df_2023_grouped['Unique Clicks'], df_2023_grouped['% of Views'], alpha=0.5, color='black', edgecolors='white')
plt.title('Correlation Between Unique Clicks and % of Views (2023)', fontsize=14, fontweight='bold')
plt.xlabel('Unique Clicks (Sum)', fontsize=12)
plt.ylabel('% of Views (Mean)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()





plt.figure(figsize=(8, 6))
sns.kdeplot(x=df_2024_grouped['Unique Clicks'], y=df_2024_grouped['% of Views'], cmap="coolwarm", fill=True, alpha=0.7)
plt.scatter(df_2024_grouped['Unique Clicks'], df_2024_grouped['% of Views'], alpha=0.5, color='black', edgecolors='white')
plt.title('Correlation Between Unique Clicks and % of Views (2024)', fontsize=14, fontweight='bold')
plt.xlabel('Unique Clicks (Sum)', fontsize=12)
plt.ylabel('% of Views (Mean)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
